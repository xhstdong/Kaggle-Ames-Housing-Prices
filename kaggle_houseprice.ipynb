### Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_filepath="E:/numerical/kaggle_prices/train.csv"  #note the back slash
test_filepath="E:/numerical/kaggle_prices/test.csv"
import os

pwd=os.getcwd()
pwd
os.chdir(os.path.dirname(train_filepath)) #change to location of data
df = pd.read_csv(os.path.basename(train_filepath),header=0) #read data

test_df = pd.read_csv(os.path.basename(test_filepath),header=0) #read dat
#df=pd.read_csv('train.csv')
os.chdir(pwd) #change back to current folder
df.head(5)
```

There is no way to know for sure, but it seems the missing values in the following column actually indicate that the amenity mentioned in the column is not present in these properties, and I have chosen a value of 0 to represent this situation:

```python
#deal with meaningful nan values (i.e. they form a category)
for col_name in ['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinSF1','BsmtFinType2','BsmtFinSF2',
             'FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']:
    df[col_name].fillna(0, inplace=True)
    test_df[col_name].fillna(0, inplace=True)
df.head(5)
```

### Other Missing Values

For other columns, it seems highly unlikely that any property would not have these amenities, and the values are truly 'missing'. In these cases, I replace numerical column with the average and categorical column with the most frequent value:

df.LotFrontage.mean()

```python
df['Electrical'].value_counts()
```

```python
for col_name in['LotFrontage']:
    df[col_name].fillna(df[col_name].mean(),inplace=True)
    test_df[col_name].fillna(test_df[col_name].mean(),inplace=True)
    
for col_name in['Electrical']:
    df[col_name].fillna(df[col_name].value_counts().idxmax(),inplace=True)
    test_df[col_name].fillna(test_df[col_name].value_counts().idxmax(),inplace=True)
```

```python
df['MasVnrType'].value_counts()
```

```python
df[df['MasVnrType']=='None'].MasVnrArea
```

```python
df['MasVnrType'].fillna('None', inplace=True)
df['MasVnrArea'].fillna(0.0, inplace=True)
df['GarageYrBlt'].fillna(3000,inplace=True) #make it somepoint in the future, hopefully wont confuse the model
test_df['MasVnrType'].fillna('None', inplace=True)
test_df['MasVnrArea'].fillna(0.0, inplace=True)
test_df['GarageYrBlt'].fillna(3000,inplace=True) #make it somepoint in the future, hopefully wont confuse the model
```

```python
df.isnull().sum()
df.drop('Id', axis=1, inplace=True)
test_id=test_df['Id']
test_df.drop('Id', axis=1, inplace=True)
#note: only uncleaned column is garage year, since they are unbuilt
```

```python
### the test set has additional missing data columns
for col_name in['MSZoning','Utilities','Exterior1st','Exterior2nd','KitchenQual','Functional','SaleType']:
    test_df[col_name].fillna(test_df[col_name].value_counts().idxmax(),inplace=True)
for col_name in ['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageCars','GarageArea']:
    test_df[col_name].fillna(0, inplace=True)
    #test_df[col_name].fillna(int(test_df[col_name].mean()), inplace=True)
test_df.isnull().sum().sum()
```

```python
#dummy_vars=pd.get_dummies(df['MSSubClass'])
#dummy_vars.columns.rename([dummy_vars.columns])
# find the 
obj_df=df.select_dtypes(include=['object'])
categorical_col=[obj_df.columns,'MSSubClass']
```

```python
#convert catgorical variables
#note: column names may differ between train and test
for col_name in categorical_col:
    dummy_vars = pd.get_dummies(df[col_name])
    dummy_vars_test=pd.get_dummies(test_df[col_name])
    # change column names for clarity: skipped, could be confusing
    #dummy_vars.rename(columns=[str(dummy_vars.columns)], inplace=True)
    #dummy_vars_test.rename(columns=[str(dummy_vars_test.columns)],inplace=True)
    
    df = pd.concat([df, dummy_vars], axis=1)
    test_df=pd.concat([test_df, dummy_vars_test], axis=1)
# drop original column "num-of-doors" from "df"
    df.drop(col_name, axis = 1, inplace=True)
    test_df.drop(col_name, axis=1, inplace=True)
df.columns=df.columns.astype(str)
test_df.columns=test_df.columns.astype(str)
test_df.head(5)
```

```python
print(df.shape)
print(test_df.shape)
```

It's very inconvenient that the columns of the train and test data sets do not match up. After some googling, I found a good solution to swith the data columns until they match.

```python
col_to_add=np.setdiff1d(df.columns, test_df.columns)

for col in col_to_add:
    test_df[col]=0
    
test_df=test_df[df.columns]
test_df.head(5)
```

```python
test_df.drop('SalePrice', axis=1, inplace=True)
print(df.shape)
print(test_df.shape)
```

### Model Building-Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import cross_val_predict, cross_val_score

x_data=df.drop('SalePrice',axis=1)
y_data=df['SalePrice']
rf=RandomForestRegressor()
#rf=RandomForestRegressor(oob_score=True)
#rf.fit(x_data, y_data)
#print(rf.oob_prediction)
scores=cross_val_score(rf,x_data,y_data, cv=10)
print( scores, np.mean(scores))
#rf.fit(x_data,y_data)
```

```python
max_score=0
max_param=[0,0]
#for estimator_num in [10]:
#    for min_leaf_num in [1, 10]:
for estimator_num in [10,30,50,70,90,110]:
    for min_leaf_num in [1,10,30,50,70,90]:
        rf=RandomForestRegressor(n_estimators=estimator_num, min_samples_leaf=min_leaf_num, 
                              n_jobs=-1, random_state=40, max_features='auto') #doesn't really understand oob yet
        score=np.mean(cross_val_score(rf,x_data,y_data, cv=10))
        print('score:', score, ' with parameters (n_estimators, min_samples_leaf): ', [estimator_num, min_leaf_num])
        if score>max_score:
            max_score=score
            max_param=[estimator_num, min_leaf_num]

print('max score: ',max_score, ' with parameters (n_estimators, min_samples_leaf): ', max_param )
```

```python
rf=RandomForestRegressor(n_estimators=max_param[0], min_samples_leaf=max_param[1], 
                              n_jobs=-1, random_state=40, max_features='auto') #doesn't really understand oob yet
   
```

```python
rf.fit(x_data,y_data)
var_importance=rf.feature_importances_
var_importance
```

```python
var_ranking=np.argsort(var_importance)[::-1]
print('feature importance in decreasing order:')
for i in var_ranking:
    print(x_data.columns[i])
```

These results are intuitive. Quality of the finish, the size of the house, and the number of highly desirable amenities such as recent construction and large garages are all important factors, while buyers typicaly don't spend time researching other factors such as electrical fuse type.
